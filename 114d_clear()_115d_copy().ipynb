```python
# Unit 114 딕셔너리의 모든 요소 삭제 : d.clear()

# d.clear() => 딕셔너리 d의 모든 요소 삭제
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
greats.clear()
print(greats)
```

```python
# del d => 딕셔너리 d 자체를 메모리에서 제거
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
del greats  # greats 자체를 메모리에서 제거 => 더 이상 존재하지 않는 변수가 됨
print(greats)  # NameError 오류
```

```python
# Unit 115 딕셔너리의 모든 요소 복제 : d.copy()

# d.copy() => 딕셔너리 d의 복사본 생성
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
buffer = greats.copy()  # greats의 복사본을 buffer에 둠 => 원본인 greats 보존 O
del buffer['관우']
print(f'greats : {greats}')
print(f'buffer : {buffer}')
```
