```python
# Unit 138 예외처리 2 : try~except~else

# try~except~else 구문 : 어떤 로직을 수행했을 때 예외가 발생하지 않았을 경우에만 특정 작업을 수행하는 코드를 작성해야 할 때 활용
try :
  잠재적으로 오류 및 예외가 발생할 수 있는 코드
except :
  예외가 발생했을 때 예외를 처리하는 코드
else :
  <try~except>로 감싸진 코드에서 예외가 발생하지 않았을 때 수행되는 코드
```

```python
# try~except로 감싼 부분을 for문 전체로 변경한 코드
fileList = ['file01.jpg', 'file02.jpg', 'file03', 'file04.jpg']
try :
  for file in fileList :
    fname = file.split('.')  # 파일 이름과 파일확장자 분리
    print(f'파일 이름 : {fname[0]}, 파일 확장자 : {fname[1]}')
except :
  print(f'ERROR!! <{file}> 처리중 오류 발생!')
else :                      # try~excpet에서 예외 발생 => else문 실행 X
  print('모든 파일을 성공적으로 처리했습니다.')
  # 'file03'을 처리 중 오류 발생 => for문의 로직 종료, except문의 print()를 출력함 -> 'file04.jpg'는 처리 X인 채 로직 종료
```

```python
# Unit 139 예외처리 3 : try~except~finally

# try~except~finally => try~except로 감싼 부분에서 예외 발생 유무와 상관없이 어떤 코드를 무조건 실행시키기 위해 이용하는 코드
try :
  잠재적으로 오류 및 예외가 발생할수 있는 코드
except :
  예외가 발생했을 때 예외를 처리하는 코드
finally :
  무조건 수행되는 코드
```

```python
# 예외가 발생하는 코드
fileList = ['file01.jpg', 'file02.jpg', 'file03', 'file04.jpg']
try :
  for file in fileList :
    fname = file.split('.')  # 파일 이름과 파일확장자 분리
    print(f'파일 이름 : {fname[0]}, 파일 확장자 : {fname[1]}')
except :
  print(f'ERROR!! <{file}> 처리중 오류 발생!')
else :
  print('모든 파일을 성공적으로 처리했습니다.')
finally :
  print('무조건 수행되는 코드!')
```
