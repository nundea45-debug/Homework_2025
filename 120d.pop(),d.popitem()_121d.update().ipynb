```python
# Unit 120 딕셔너리에서 요소를 제거한 후 값 또는 요소 얻기 : d.pop(), d.popitem()
```

```python
# d.pop()

# d.pop(key) => 딕셔너리 d에 key가 있음 -> key에 해당하는 요소 삭제 후 대응값을 리턴
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
level = greats.pop('관우')  # greats에서 '관우'가 키인 요소 삭제 후 대응값 리턴 ('관우' 대응값 -> 95 => level의 값 -> 95)
print(f'greats : {greats}\nlevel : {level}')  # greats에서 '관우':95가 삭제된 값 출력
```

```python
# 딕셔너리 d에서 key가 없을 시 => KeyError 오류 발생
level = greats.pop('제갈량')
```

```python
# d.pop(key, default) => 딕셔너리 d에 key가 있을 시 -> key에 해당하는 요소 삭제 / key가 없을 시 -> default 리턴
level = greats.pop('제갈량', 99)  # greats에 '제갈량'이 키로 존재 X => 디폴트 값으로 지정한 99를 리턴
print(f'greats : {greats}\n제갈량의 레벨 : {level}')
```

```python
# d.popitem()

# d.popitem() => 딕셔너리 d에서 맨 마지막으로 추가한 요소 제거 -> 삭제된 요소를 튜플 (key, val)로 리턴
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
while len(greats) > 0 :  # greats가 빈 딕셔너리가 될 때까지 while문을 실행
  general = greats.popitem()  # greats.popitem() => greats의 맨 마지막 요소를 제커 후 해당 요소를 튜플로 리턴
  print(f'추출한 장군 : {general}\ngreats 상태 : {greats}')
  print('-'*50)
```

```python
# 빈 딕셔너리에서 popitem() 호출 시 => KeyError 발생
general = greats.popitem()
```

```python
# Unit 121 딕셔너리 갱신하기 1 : d.update()

# d.update(d1) => 딕셔너리 d에 딕셔너리 d1을 통합한 딕셔너리 d를 만듦 (d1에 존재하는 키가 d에도 존재 시 => d1의 요소로 덮음)
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
others = {'제갈량':99, '황충':70}
greats.update(others)  # greats에 others를 통합해 갱신 (greats.update(제갈량=99, 황충=70)으로 대체 가능)
print(greats)
```

```python
greats.update(유비=90, 관우=96)  # greats에 '유비', '관우'가 키로 존재 => '유비'의 값을 90, '관우'의 값을 96으로 덮어 씀 (문자열이 키인 경우 => '', ""를 사용 X)
print(greats)
```
