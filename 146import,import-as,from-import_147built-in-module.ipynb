```python
# Unit 146 모듈 임포트 : import, import~as, from~import

# import
# 모듈을 임포트하는 일반적인 방법
import 모듈 이름             # 모듈 => 내장 모듈과 외부 모듈이 존재
import 패키지 계층.모듈 이름 # 패키지 계층 => 모듈이 있는 곳까지 '.'으로 구분된 계층구조
```

```python
# 파이선 내장 모듈 => 모듈을 따로 설치할 필요 X -> import 키워드를 이용해 해당 모듈을 임포트하여 활용
import math                             # 수학 관련 파이썬 내장 모듈인 math를 임포트
print(f'sqrt(2) = {math.sqrt(2):.3f}')  # math 모듈에서 제곱근을 계산하는 함수인 sqrt()를 이용 -> 2의 제곱근을 소수점 3자리까지 출력
```

```python
# import~as
import 모듈 as 별명  # 모듈의 이름이 길거나 복잡 => 임포트할 모듈의 별명을 지정
```

```python
import math as m                    # math 별명을 m으로 지정 => math 모듈은 m으로 참조 가능
print(f'sqrt(2) = {m.sqrt(2):3f}')  # m.sqrt(2)는 math.sqrt(2)와 같음
```

```python
# from~import~
# 모듈에서 원하느 함수만, 패키지에서 원하는 모듈만 특정하여 임포트하는 방법
from 모듈 import 함수1, 함수2, ...
from 패키지 import 모듈1, 모듈2, ...
from 패키지 계층 import 함수1, 함수2, ...
```

```python
from math import sqrt, log         # math 모듈에서 sqrt, log 함수만 임포트
print(f'sqrt(2) = {sqrt(2):.3f}')  # sqrt(2), log(2)와 같이 모듈 이름 없이 임포트한 함수 이름 그대로 사용
print(f'log(2) = {log(2):.3f}')
```

```python
# 임포트한 모듈을 찾을 수 X => ModuleNotFoundError 예외 발생
import mymodule
```

```python
# Unit 파이썬 내장 모듈(built-in module) vs. 외부 모듈

# 내장 모듈(built-in module) => 사용자가 따로 설치할 필요 없이 파이썬에 내장된 모듈, 매우 많이 존재
# 외부 모듈 => 파이썬 내장 모듈 이외의 모듈들은 외부 모듈, 필요시 사용자가 따로 설치해야 모듈에 포함된 함수 사용 가능
```

```python
# 사용자가 따로 설치하지 않은 외부 모듈을 임포트 => ModuleNotFoundError 오류 발생
import matplotlib # matplotlib 모듈을 설치 후 import해 해결 (외부 모듈 설치 => pip 이용)
```
