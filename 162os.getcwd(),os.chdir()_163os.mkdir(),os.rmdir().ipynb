```python
# Unit 162 현재 작업 디렉토리 확인, 작업 디렉토리 변경 : os.getcwd(), os.chdir()

# os.getcwd() => 현재 작업 디렉토리의 절대경로를 리턴
from os import getcwd

print(f'현재 작업 디렉토리 : {getcwd()}')
```

```python
# os.chdir(path) => path로 지정된 디렉토리로 이동(path는 절대경로나 상대경로로 지정 O)

from os import chdir

print(f'현재 작업 디렉토리 : {getcwd()}')
chdir('..')                # 상대경로 지정  # 현재 작업 디렉토리의 상위 디렉토리로 작업 디렉토리를 이동
print(f'현재 작업 디렉토리 : {getcwd()}')
chdir('c;/mywork/jupyter') # 절대경로 지정  # 절대경로로 지정된 디렉토리로 작업 디렉토리를 이동
print(f'현재 작업 디렉토리 : {getcwd()}')
```

```python
# Unit 163 디렉토리 생성 및 제거 : os.mkdir(), os.rmdir()

# os.mkdir(folder) => folder로 지정된 절대경로 또는 상대경로의 디렉토리를 생성
from os import mkdir

folder = 'myfolder'
mkdir(folder)  # 현재 작업 디렉토리에서 'myfolder'라는 이름의 디렉토리 생성
```
