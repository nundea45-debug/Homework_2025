```python
# Unit 156 텍스트 파일 저장하기 : f.writelines()

# f.writelines(data) => 텍스트 data를 파일에 저장(data => 유니코드 문자열이나 문자열이 요소인 리스트(data가 리스트인 경우 -> 요소를 순서대로 파일에 저장))
```

```python
data1 = 'This is data1'
data2 = 'This is data2'
with open('mydata.txt', 'w') as f :  # data1과 data2를 f.writelines()를 이용해 파일에 저장
  f.writelines(data1)
  f.writelines(data2)
```

```python
with open('mydata.txt', 'r') as f :
  data = f.read()
print(data)  # data1과 data2가 연결되어 저장됨
```

```python
# data1과 data2를 각각 라인 단위 단위로 저장
data1 = 'This is data1\n'
data2 = 'This is data2\n'
```

```python
data = ['This is data1\n', 'This is data2\n']  # data => 문자열이 요소인 리스트
with open('mydata.txt', 'w') as f :            # f.writelines(data) => data의 모든 요소를 순서대로 파일에 저장
  f.writelines(data)
```

```python
# mydata.txt 파일을 읽기 모드로 열고 저장된 내용 출력
with open('mydata.txt', 'r') as f :
  data = f.read()
print(data)
```

```python
# Unit 157 바이너리 파일 복사하기 : f.read(), f.write()

bufsize = 1024  # 1K 바이트  # bufsize를 1024바이트(= 1KB)로 정의
with open('sunrise.jpg', 'rb') as f :  # sunrise.jpg를 바이너리 읽기 모드로 오픈
  with open('sunrise.jpg', 'wb') as h :# sunrise_copies.jpg를 바이너리 쓰기 모드로 오픈
    data = f.read(bufsize)  # 'sunrise.jpg'를 1KB씩 읽어 들임  # sunrise.jpg의 데이터를 1024바이트만큼 읽고 data로 둠
    while data :  # data의 내용이 있을 때까지 while문을 반복
      h.write(data)  # 읽은 1KB 데이터를 씀  # 읽은 data를 sunrise_copied.jpg에 씀 -> 다시 sunrise.jpg의 데이터 1024바이트를 읽고 data로 둠
      data = f.read(bufsize)  # 'sunrise.jpg'를 1KB씩 읽어 들임  # data의 내용이 있을 시 => 6라인 수행
```
