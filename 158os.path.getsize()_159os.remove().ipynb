```python
# Unit 158 파일 크기 구하기 : os.path.getsize()

# os.path.getsize(file) => os.path 모듈의 getsize(file) -> file의 크기를 바이트 단위로 리턴(file => 파일명을 포함한 상대경로나 절대경로)
```

```python
from os.path import getsize          # os.path 모듈의 getsize()를 임포트

file1 = 'sunrise.jpg'                # 파일 이름만
file2 = 'C:/data/temp/DJI_0128.MP4'  # 절대경로
file3 = '../../temp/DJI_0129.MP4'    # 상대경로
```

```python
print(f'{file1} : {getsize(file1):,} 바이트')  # getsize(file1) => 코드가 실행되는 티렉토리에서 file1을 찾아 크기를 리턴
print(f'{file2} : {getsize(file2):,} 바이트')  # getsize(file2) => 절대경로로 지정된 파일의 크기를 리턴
print(f'{file3} : {getsize(file3):,} 바이트')  # getsize(file3) => 상대경로로 지정된 파일의 크기를 리턴
# getsize()의 인자로 입력된 파일 이름이나 파일경로에 해당 파일 존재 X => FileNotFoundError
```

```python
# Unit 159 파일 삭제하기 : os.remove()

# os.remove(file) => os 모듈의 remove(file) -> file을 삭제(file => 파일명을 포함한 상대경로나 절대경로), 파일을 삭제할 것인지 묻지 X -> 바로 삭제
```

```python
from os import remove

target = 'mydata.txt'
k = input(f'<{target}>을 삭제하시겠습니까? (y/n)')
if k.lower() == 'y' :
  remove(target)
  print(f'<{target}>을 삭제했습니다.')
else :
  print(f'<{target}>을 삭제하지 않았습니다.')
```
