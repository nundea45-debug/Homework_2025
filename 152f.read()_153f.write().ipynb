```python
# Unit 152 파일 읽기 : f.read()

# f.read(size) => 파일 f의 내용을 size만큼 읽으라는 의미(size 단위 => 바이트(byte, size 생략 -> 파일객체 f의 모든 내용을 읽음)
```

```python
# 텍스트 파일 읽기 1(영어)
with open('prophet_en.txt', 'r') as f :  # prophet_en.txt 텍스트 파일을 읽기 모드, UTF-8로 오픈
  readData = f.read()                    # prophet_en.txt 파일의 모든 내용을 읽어서 유니코드 문자열로 리턴
print(readData)
```

```python
# 텍스트 파일 읽기 2(한글)
with open('prophet_ko.txt', 'r') as f :
  readData = f.read()
print(readData)
```

```python
# 텍스트 파일 읽기 3(CP949 인코딩)
with open('prophet_cp949_ko.txt', 'r') as f:  # 파일 오픈 시 encoding 인자 지정 => 디폴트 인코딩 방식이 UTF-8로 설정
  readData = f.read()                         # UnicodeDecodeError 예외 발생(CP949로 인코팅된 파일을 UTF-8로 읽으려고 했기 때문)
print(readData)
```

```python
# 텍스트 파일 읽기3(CP949 인코팅 수정 코드)
with open('prophet_cp949_ko.txt', 'r', encoding = 'cp949') as f :
    readData = f.read()
print(readData)
```

```python
# 바이너리 파일 읽기
with open('sunrise.jpg', 'rb') as f :  # 파일을 'rb'로 오픈(바이너리 읽기 모드)
  readData = f.read(50)                # sunrise.jpg 파일의 첫 50 byte를 읽음 -> 읽은 값은 byte 문자열로 리턴
print(readData.hex('-'))               # byte 문자열 readData를 '-'로 구분된 1byte 단위 16진수로 출력
```

```python
# Unit 153 파일 쓰기 : f.write()

# f.write(data) => 파일 f에 data를 저장(data가 텍스트 일 경우 -> f를 오픈할 때 저장한 인코딩 방식을 적용해 저장)
```

```python
# 텍스트 파일에 내용 쓰기
data = input('저장할 문장을 입력하세요 :')  # 사용자가 입력한 문장을 받음
flag = input('입력한 문장을 저장하시겠습니까? (Y/N)')
if flag.lower() == 'y' :
  with open('mydata.txt', 'w') as f :      # 파일을 텍스트 모드, UTF-8 인코딩 방식으로 오픈(파일이 X -> 파일을 생성/ 파일 O -> 기존 파일 삭제 후 생성 )
    f.write(data)                          # 파일 오픈 시 지정한 인코딩 방식으로 data를 인코딩해서 저장
  print('입력한 문장을 "mydata.txt"에 저장했습니다.')
else :
  print('입력한 문장을 저장하지 않습니다.')
```

```python
# 바이너리 파일에 내용 쓰기
data = b'A lot of things occur each day'  # byte문자열 data를 정의
with open('mydata.bin', 'wb') as f :      # mydata.bin 파일을 바이너리 쓰기 모드로 오픈
  f.write(data)                           # byte문자열 data를 mydata.bin에 저장
```

```python
# 저장된 파일 내용을 읽기
with open('mydata.bin', 'rb') as f :  # mydata.bin은 바이너리 읽기 모드로 오픈
  data = f.read()                     # 파일에 저장된 데이터를 읽어 data로 둠
print(data)                           # data를 byte문자열로 출력
```
