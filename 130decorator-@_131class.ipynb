```python
# 출력물의 아래 위에 점선으로 구분할 수 있도록 함 => report(), makeNumber() 수정
def report1() :
  print('-'*50)            # 출력물의 위쪽 점선
  print('이 부분은 리포트의 본문입니다.')
  print('-'*50)            # 출력물의 아래쪽 점선
  return

def makeNumber1() :
  print('-'*50)            # 출력물의 위쪽 점선
  for i in range(3) :
    print(i)
  print('-'*50)            # 출력물의 아래쪽 점선
  return
```

```python
# print('-'*50) 부분을 호출하는 함수의 종류와 무관하게 모두 적용할 수 있는 방법
def makeForm(func) :  # 인자로 함수를 입력받음
  def drawLine() :    # drawLine()을 정의
    print('-'*50)
    func()            # drawLine() => 인자로 입력받은 func() 실행 전후로 print('-'*50)을 실행하는 함수
    print('-'*50)
  return drawLine     # draWLine() 함수를 리턴
```

```python
report_new = makeForm(report)  # makeForm() 안에서 정의된 drawLine을 리턴 (func()는 makeForm()의 인자로 전달받은 report가 됨)
report_new()                   # report_new => func이 report인 drawLine임
```

```python
# 위의 2라인 report_new()를 실행 => 아래 함수를 실행
def drawLine() :
  print('-'*50)
  report()
  print('-'*50)
```

```python
# makeNumber()에 대해서도 적용
makeNumber_new = makeForm(makeNumber)  # makeForm(func) => 인자로 입력된 모든 함수에 대해 실행 전후 점선을 출력하는 기능 O
makeNumber_new()
```

```python
# 데코레이터 설명
@makeForm  # 이 부분이 데코레이터  # 해석 : "바로 아래에 정의된 함수 report()를 실행할 때, 데코레이터 함수 makeForm(func)의 인자로 report를 입력해서 리턴받은 함수를 실행하시오"
def report() :
  print('이 부분은 리포트의 본문입니다.')
  return
```

```python
# 해석한 문구를 코드로 풀어 씀
# report()를 실행하면 아래의 코드와 같이 수행하세요
ret = makeForm(report)
ret()  # report()와 makeNumber()의 실행 전후로 점선을 그리는 로직
```

```python
# 데코레이터 @makeForm 없이 정의된 report()에 대한 코드 결과는 데코레이터 @makeForm과 함께 정의된 report()를 단순 실행한 결과와 동일
report_new = makeForm(report)
report_new()
```

```python
# makeNumber()에 대한 데코레이터 추가
@makeForm
def makeNumber() :
  for i in range(3) :
    print(i)
  return

ret = makeForm(makeNumber)
ret()

makeNumber_new = makeForm(makeNumber)
makeNumber_new()

makeNumber()
```

```python
# 데코레이터와 함께 정의된 함수에 인자가 있을 경우, 데코레이터 함수 정의
def decoFunc(func) :
  def wrapper(*args, **kwargs) :
    # 공통 기능 수행 코드
    func(*args, **kwargs)
    # 공통 기능 수행 코드
  return wrapper
```

```python
# Unit 131 클래스(Class) 개념

# 클래스(Class) => 지정된 이름으로 만들어진 하나의 독립 공간 (속성을 담을 변수, 득정 기능을 수행하는 함수들의 집합으로 구성)
# 매서드(method) => 클래스 안에서 정의되는 함수
# 클래스 정의
class 클래스 이름 :
  속성을 담기 위한 변수 정의
  매서드 정의
```

```python
# 클래스를 정의하고 객체를 생성
class Wizard :
  num_of_generated = 0
  def setName(self, name) :  # setName(self, name) => 클래스 안에서 정의된 매서드 (매서드 정의 => 일반 함수처럼 def 키워드로 정의)
    self.name = name
```

```python
# 클래스 객체 생성 방법
객체이름 = 클래스 이름(인자1, 인자2, ...)  # 클래스 이름() 안의 인자 => 클래스 생성자로 입력되는 인자
```

```python
player1 = Wizard()             # Wizard 클래스를 인스턴스화 한 객체 생성 후 player1이라는 이름으로 둠 (player1은 Wizard 클래스의 객체가 됨)
player1.setName('FirePlayer')  # Wizard 객체의 setName()을 실행 (setName()은 객체 내의 인스턴스 변수 self.name에 'FirePlayer'로 설정)
print(player1.name)            # Wizard 객체 player1의 인스턴스 변수 self.name 값 출력 (=> 'FirePlayer'가 출력)
```
