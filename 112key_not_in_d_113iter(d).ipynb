```python
# Unit 112 딕셔너리에 특정 키가 없는지 확인 : key not in d

# key not in d => key가 d의 키로 존재하지 X => True
```

```python
frequency = {}
while True :
  c = input('알파벳 1자를 입력하세요[끝내려면 엔터를 치세요]:')
  if c == '':
    break
  if c not in frequency :  # frequency에 사용자가 입력한 문자인 c가 키로 존재 X => True
    frequency[c] = 1       # frequency에 c : 1을 요소로 추가
  else :
    frequency[c] += 1      # frequency에 c가 키로 존재 => c에 대응하는 값에 1을 더함
  print(frequency)
```

```python
# Unit 113 디셔너리의 모든 키로 이터레이터 생성 : iter(d)

# iter() => 딕셔너리의 키로 이루어진 이터레이터 생성
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
for name in iter(greats):  # 딕셔너리 자료 greats의 키로 이루어진 이터레이터 생성
  print(name)
# irer(d) == iter(d.keys())와 동일한 이터레이터 생성
```
