```python
# Unit 144 모듈(module) 이해하기

# 모듈(module) => 이미 만들어져 있고 안정성이 검증된 함수들을 성격에 맞게 분류해서 하나의 파이썬 파일(확장자 = .py)에 묶어 놓은 것
파이썬 모듈 = 모듈 이름.py
# 모듈을 임포트(inport)하다. => 모듈을코드로 가져와서 자유롭게 사용할 수 있도록 함(모듈에 있는 함수를 활용하기 위해)
```

```python
def splitFilename(file) :  # slitFilename(file) => 인자로 입력된 파일 이름을 순수 파일 이름으로 확장자로 분리한 결과를 튜플로 리턴하는 함수
  buf = file.split('.')
  if len(buf) > 1 :
    ext = buf[-1]
    fname = '.'.join(buf[:-1])
  else :
    fname = buf[0]
    ext = None
  return (fname, ext)
```

```python
#import splitfilename   # splitfilename 모듈을 임포트

filename = 'python.200,study.py'
ret = splitFilename(filename)  # 모듈에 있는 함수 활용을 위한 호출 (모듈 이름.함수)
print(f'파일 이름 : {ret[0]}, 파일확장자 : {ret[1]}')
```
