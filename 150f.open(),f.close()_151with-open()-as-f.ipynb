```python
# 파일 오픈 시 모드와 인코팅 방식 모두 생략 => 디폴트 값(텍스트 읽기 모드)으로 파일 오픈
파일 객체 = open(파일경로)  # 텍스트 읽기 모드로 파일을 오픈
```

```python
# 일반적인 로직 => 파일 오픈 -> 관련 작업 수행 -> 파일 클로즈
# 텍스트 파일을 'cp949' 인코딩으로 읽거나 쓰기
f = open('myfile.txt', 'w', encoding = 'cp949')
<파일 처리 코드>
f.close()

# 바이너리 이미지 파일 읽기
f = open('image.jpg', 'rb')
<파일 처리 코드>
f.close()
```

```python
# 파일을 열고 자동으로 닫기 : with open() as f

# 파일 오픈 -> 관련 작업 수행 -> 자동으로 파일 클로즈
with open(파일경로, 모드, encoding = 인코딩 방식) as 파일객체 :
  <파일 처리 코드>  # <파일 처리 코드> 종료 => 자동으로 오픈한 파일 클로즈
```

```python
f = open('myfile.txt', 'r')
readData = f.readlines()
f.close()
```

```python
# 위 코드와 동일한 기능 수행
with open('myfile.txt', 'r') as f :
  readData = f.readlines()
```
