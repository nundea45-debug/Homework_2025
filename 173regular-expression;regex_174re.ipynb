```python
# 반복 표현식 예시
ab?c : a와 c사이에 b가 0 또는 1회 나타나는 문자열과 매치됨          # 매치되는 문자열 : ac, abc
ab*c : a와 c사이에 b가 0회 이상 반복으로 나타나는 문자열과 매치됨   # 매치되는 문자열 : ac, abc, abbbc 등
ab+c : a와 c사이에 b가 1회 이상 반복으로 나타나는 문자열과 매치됨   # 매치되는 문자열 : ac를 제외하면 ab*c와 동일
ab{3}c : a와 c사이에 b가 3회 반복으로 나타나는 문자열과 매치됨      # 매치되는 문자열 : abbbc
ab{1, 3}c : a와 c사이에 b가 1~3회 반복으로 나타나는 문자열과 매치됨 # 매치되는 문자열 : abc, abbc, abbbc
```

```python
# 매칭 표현식 예시
1.1 : 1 + (한글자로 된 모든 문자) + 1인 문자열과 매치됨  # 매치되는 문자열 : 1a1, 1%1, 111 ... 등
<.*> : <>로 둘러싸인 모든 문자열과 매치됨                # 매치되는 문자열 : <>, <1>, <ok>, <href = /> ... 등
^\d.* : 숫자로 시작하는 모든 문자열과 매치됨             # 매치되는 문자열 : 24k, 77Lover ... 등
^\d.+\d$ : 숫자로 시작하고 숫자로 끝나는 문자열과 매치됨 # 매치되는 문자열 : 1python33, 0loverly87 ...  등
\d{6}-\d{7} : 6개의 숫자 -7개의 숫자 인 문자열과 매치됨  # 매치되는 문자열 : 주민번호 형식, 123456-1234567 ... 등
```

```python
# Unit 174 정규표현식 응용

# re => 정규식과 관련된 다양한 기능을 제공
# re 모듈에서 자주 활용되는 메서드 4개(compile, match, search, findall)
```

```python
# re.compile[]
p = re.compile(pattern, flags=0)  # pattern => 정규식 객체를 리턴 / flags => 정규식 컴파일 옵션, "re.I"로 설정하고 컴파일(대소문자 구분 X, 매칭 수행 지시)
```

```python
import re
p = re.compile(r'href=".*"', re.Ⅰ)  # 정규식 "href=".*""를 컴파일한 정규식 객체를 리턴 # r => 문자열이 raw string임을 의미 # re.i => 매칭 수행 시 대소문자 구분 X
```

```python
# match[]
p.match(contents)
re,match(pattern, contents, flags=0) # contents의 첫 부분부터 정규식 pattern과 일치하는지 검토 -> 매칭 성공 => Match 객체 리턴 / 매칭 실패 => None을 리턴
```

```python
contents = """<a href = "help.html">도움말을 보려면 여기를 클릭하세요</a><font size="15">"""
ret = p.match(contents)  # 패턴 p가 contents의 첫 부분부터 일치하는지 검토(패턴 p => 'href=".*"' /= contents => '<a hret로 시작 ==> 매칭 실패)
print(ret)
```

```python
contents2 = """href = "help.html">도움말을 보려면 여기를 클릭하세요</a><font size="15">"""  # <a를 삭제해서 의도적으로 패턴 p와 매칭에 성공하도록 만든 문자열
ret = p.match(contents2)
print(ret.group())  # 매치된 문자열을 리턴 -> 매치된 문자열 출력
```

```python
# search[]
p.search(contents)
re.search(pattern, contents, flags = 0)  # content 처음부터 전제를 탐색하며 정규식 pattern과 최초로 매칭되는 결과를 Match 객체로 리턴(매칭 실패 => None 리턴)
```

```python
ret = p.search(contents)  # contents에서 p와 최초로 매칭되는 문자열을 Match 객체로 리턴
print(f'Match.group()={ret.group()}')  # 패턴이 매치된 Match 객체의 메서드의 리턴값을 출력(2~5열)
print(f'Match.start()={ret.start()}')
print(f'Match.end()={ret.end()}')
print(f'Match.span()={ret.span()}')
```

```python
# findall[]
p.findall(contents)
re.findall(pattern, contents, flags=0)  # contents의 처음부터 전체를 탐색하며 정규식 pattern과 일치라는 모든 문자열을 리스트로 리턴(매칭 실패 => 빈 리스트([]) 리턴)
```

```python
ret = p.findall(contents)  # contents에서 p와 매칭되는 모든 문자열을 요소로 하는 리스트를 리턴
print(ret) # 매칭되는 결과가 1개 => 문자열 1개가 요소인 리스트 출력
```
