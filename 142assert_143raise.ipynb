```python
# Unit 142 예외 발생시키기 1 : assert

# assert문 => assert 뒤에 놓이는 조건이 참이 X -> AssertionError 예외를 발생
assert 조건, 예외 메시지
# assert => 조건이 거짓 -> 예외 메시지와 함께 AssertionError 오류 발생 (주어진 조건이 참임을 보증하고자 할 때 사용)
```

```python
# 'assert 조건, 예외 메시지'
if __debug__:  # __debug__ => 값은 파이썬 내장 변수 True, 인위적으로 변경 X
  if not 조건 : raise AssertionError(예외 메시지)
```

```python
# 파이썬 실행 시 __debug__의 값을 인자 -0으로 설정
python -0 pythontest.py
```

```python
data = ['a', 'bc', 1234, 'def', 'ghij']
try :
  for d in data :
    assert type(d) is str, f'문자열이 아닌 데이터 <{d}> 있음!'  # 'type(d) is str' => d의 자료형이 문자열 = True, 문자열 X = False / d의 값이 정수 1234 => AssertionError 예외 발생
    print(d)
except AssertionError as e :  # AssertionError 예외 발생 => for문 실행 종료, except문에서 예외 처리
  print(e)
```

```python
# Unit 143 예외 발생시키기 2 : raise

# 코드에서 특정 예외를 강제로 발생시키는 방법
raise 예외클래스('프로그래머가 지정한 예외 메시지')
# raise문 => 지정된 예외 클래스를 강제로 발생 (예외클래스 -> 파이썬이 지원하는 것 이어야하며 프로그래머가 지정한 메시지임)
```

```python
# RuntimeError를 발생시키는 함수 정의
def makeException() :
  raise RuntimeError('뭔가 잘못되었음')
```

```python
# 위 함수를 호출하는 코드를 실행
try :
  makeException()
except Exception as e :
  print(e)
```

```python
# 코드에 raise를 추가 후 실행 (except문 안에서 raise를 단독 호출 시 => 현재 처리 중인 오류를 다시 발생)
try :
  makeException()
except Exception as e :
  print(e)
  raise
```

```python
# except문 안에서 특정 예외 강제 발생
try :
  makeException()  # raise RuntimeError('뭔가 잘못되었음') 예외를 강제 발생
except Exception as e :  # RuntimeError 에외처리를 위해 except문의 예외처리 코드 수행 => ValueError('값이 잘못된 것 같음') 예외를 강제 발생
  raise ValueError('값이 잘못된 것 같음')  # 강제 발생시킨 ValueError('값이 잘못된 것 같음')에 대한 예외처리 코드 존재 X => 위와 같은 출력 결과가 나옴
```
