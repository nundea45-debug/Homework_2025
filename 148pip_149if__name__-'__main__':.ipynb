```python
# Unit 149 if__name__=='__main__' :

# __name__ => 파이썬 변수, 파이썬 인터프리터가 코드를 실행하거나 임포트하기 위해 코드를 읽는 시점에 해당 변수에 값을 채움
# '__main__' => 파이썬 인터프리터가 첫 번째로 실행하는 모듈을 의미(프로그램이 구동되는 진입점)
```

```python
def printName() :
  print(f'module_b의 __name__ = {__name__}')
```

```python
from module_b import printName  # 임포트되는 module_b의 __name__ => 모듈 이름인 'module_b'로 설정됨
                                # 실행되는 코드의 __name__ => '__main__'으로 설정됨
printName()
print(f'실행 모듈인 이 모듈의 __name__ = {__name__}')
```

```python
# if __name__ == '__main__' : (모듈의 __name__이 '__main__'과 같으면) => 처음으로 실행되는 모듈이면 다음 코드를 실행하라
def printName() :
  print(f'module_b의 __name__ = {__name__}')

if __name__ == '__main__' :  # True가 되어 if문이 실행 (module_b의 __name__이 '__main__'으로 설정됨)
  print('이 모듈은 처음 실행되는 코드입니다.')
  printName()
```
