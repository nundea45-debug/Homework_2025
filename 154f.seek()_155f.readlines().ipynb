```python
# Unit 154 파일의 특정 부분만 읽고 복사 : f.seek()

# f.seek(offset) => 파일 f에서 데이터를 읽고 쓰기 위한 파일포인터의 위치를 offset으로 설정(offset의 단위 => byte)
```

```python
with open('prophet_en.txt', 'r') as f :
  f.seek(335)      # 파일포인터를 335바이트 위치로 이동
  data = f.read()  # 파일포인터가 위치한 335바이트부터 파일의 끝까지 읽음
print(data)
```

```python
# Unit 155 텍스트 파일 한 라인씩 읽기 : f.readline(), f.readlines()

# f.readline() => 텍스트 파일에서 한 라인을 읽음 -> 파일포인터를 그 다음 라인의 처음으로 이동
# 더 읽을 라인 존재 X => f.readline()은 빈 문자열을 리턴
```

```python
with open('prophet_en.txt', 'r') as f :
  line = f.readline()        # prophet_en.txt 파일 내용의 한 라인을 읽음
  while line != '' :         # line이 빈 문자열일 때까지 while문을 반복
    print(f'>>{line[:-1]}')  # f.readline()으로 읽은 라인의 맨 마지막 문자 => 줄바꿈 문자('\n') / line[:-1] => '\n'을 제외한 라인 내용
    k = input('다음 줄 : 엔터 입력. 종료 :/q 입력 :')
    if k == '/q' :           # 엔터 입력 => 다음 라인을 읽음 / '/q'를 입력 => while문 탈출 후 종료
      break
    line = f.readline()
```

```python
# f.readlines() => 텍스트 파일의 모든 라인을 읽음 -> 각 라인을 요소로 하는 리스트를 리턴
with open('prophet_en.txt', 'r') as f :
  lines = f.readlines()      # f.readlines() => prophet_en.txt의 모든 라인을 읽음 -> 각 라인을 요소로 하는 리스트를 리턴
  for line in lines :        # for문을 이용해 lines의 각 요소를 출력
    print(f'>.{line[:-1]}')
    k = input('다음 줄 : 엔터 입력. 종료 : /q 입력 :')
    if k == '/q' :
      break
```
