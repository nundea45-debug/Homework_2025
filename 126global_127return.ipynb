```python
# Unit 126 지역변수(local variable)와 전역변수(global variable) : global

# 변수 => 지역변수(local variable)와 전역변수(global variable)로 구분 (변수가 미치는 유효 범위를 기준)
# 지역변수(local variable) => 함수 안에서 정의되어 활용되는 변수 (정의된 함수 안에서만 유효한 변수)
# 전역변수(global variable) => 함수 바깥에서 정의된 변수 (코드 전반에 걸쳐 유효한 변수)
```

```python
param = 10  # param => 전역변수 (함수 바깥에서 정의)
def func1(param) :  # localData  => 지역변수 (param과 func1()안에서 정의)
  localData = 5     # param과 locaData => 매개변수 (func1() 바깥에서는 유효한 변수 X)
  print(f'param = {param}')  # "param = 5"를 출력
  return

func1(5)  # func1(5) 호출 => func1(param)의 매개변수 param = 5가 됨
print(f'param = {param}')  # func1() 바깥에서 실행되는 코드 (param => "param = 10"을 출력 (전역변수로 선언된 param))
```

```python
param = 10
def func2() :  # 지역변수 param = 20으로 설정하고 이 값을 리턴하는 함수
  param = 20   # param => 지역변수 param으로 취급 (전역변수 param에 영향 X)
  return param

ret = func2() # ret의 값은 20입니다.
print(f'param = {param}')  # {param} => 전역변수 param의 값인 10이 됨(지역변수 param은 함수 바깥에서는 유효하지 X)
```

```python
param = 10
def func3() :
  global param  # 전역변수 param임을 명시함
  param = 20    # 전역변수 param의 값을 20으로 바꿈
  return

func3()  # func3()을 호출
print(f'param = {param}')  # 전역변수 param의 값을 출력
```

```python
# Unit 127 함수 처리 결과 리턴 : return

# return => 함수 처리 결과를 함수를 호출한 곳으로 반환하는 키워드 (모든 함수의 처리 결과를 리턴 / 자료형에 제한 X)
def reverse(*args) :
  result = []
  idx = len(args)-1
  while idx >= 0 :
    result.append(args[idx])
    idx -= 1
  return result  # args로 전달받은 인자를 역순으로 저장한 result를 리턴
```

```python
ret = reverse('가', '나', '다', '라')
print(ret)
```
