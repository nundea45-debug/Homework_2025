```python
# Unit 164 하위 디렉토리 및 파일 전체 삭제 : shutil.rmtree()

# shutil.rmtree(folder) => folder로 지정된 절대경로 또는 상대경로의 디렉토리와 그 하위 디렉토리 및 모든 파일들을 일괄적으로 삭제(folder에 포함된 모든 파일과 디렉토리가 제거)
```

```python
from shutil import rmtree

folder = 'c:/mywork/jupyter/myfolder'
k = input(f'<{folder}>를 완전히 삭제하시겠습니까? (y/n) : ')
if k.lower() == 'y' :
  rmtree(folder)  # folder로 저장된 'c:/mywork/jupyter/myfolder' 디렉토리와 그 하위 디렉토리 및 모든 파일들을 전부 삭제
  print(f'<{folder}>를 완전히 삭제했습니다.')
else :
  print(f'<{folder}>를 삭제하지 않습니다.')
```

```python
# Unit 165 파일이 존재하는지 체크 : os.path.exists()

# os.path.exists(target) => target이 존재 -> True / 존재 X -> False를 리턴 (target => 파일이나 디렉토리에 대한 절대경로, 상대경로)
```

```python
from os.path import exists

print(exists('sunrise.jpg'))               # 현재 작업 디렉토리에서 sunrise.jpg가 있는지 확인
print(exists('tmp'))                       # 현재 작업 디렉토리에서 tmp라는 디렉토리가 있는지 확인
print(exists('c:/mywork/jupyter/bts.mp4')) # 절대경로로 주어진 파일 bts.mp4가 있는지 확인
```

```python
# 사용자가 입력한 이름으로 디렉토리를 만드는 코드 (이미 입력한 이름의 디렉토리 존재 -> 사용자 입력을 다시 유도)
from os import mkdir
from os.path import exists

while True :
  folder = input('생성할 디렉토리 이름을 입력하세요 : ')
  if exists(folder) :  # 사용자가 입력한 디렉토리 이름이 이미 존재 => 메시지 출력 + while문을 다시 반복
    print(f'<{folder}>가 이미 존재합니다.')
    continue
  else :               # 사용자가 입력한 디렉토리 이름으로 된 디렉토리가 X => 디렉토리 생성 + 메시지 출력 + while문 탈출
    mkdir(folder)
    print(f'<{folder}>를 생성했습니다.')
    break
```
