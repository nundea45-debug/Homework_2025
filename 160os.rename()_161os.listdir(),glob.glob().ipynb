```python
# Unit 160 파일 이름 변경 / 파일 이동 : os.remove()

# os.remove(src, dat) => src를 dst로 이동(src와 dst가 동일 디렉토리에 있을 시 -> 단순히 파일 이름만 변경)
```

```python
from os import rename

src = 'sunrise_copied.jpg'
dst = 'sunrise2.jpg'
rename(src, dst)  # sunrise_copied.jpg를 sunrise2.jpg로 변경(dst로 설정한 이름이 이미 존재 -> FileExistError)
```

```python
# Unit 161 디렉토리에 존재하는 파일 목록 얻기 : os.listdir(), glob.glob()

# os.listdir(folder) => folder에 존재하는 모든 파일과 디렉토리를 리스트로 리턴(folder는 절대경로, 상대경로로 지정 O)
```

```python
from os import listdir

folder = './'  # 현재 작업 디렉토리  # 코드가 실행되는 현 디렉토리를 folder로 지정
ret = listdir(folder)               # folder에 존재하는 모든 파일과 디렉토리를 리스트로 출력
print(ret)
```

```python
# glob.glob(target) => target에 해당하는 파일이나 디렉토리를 리스트로 리턴(target은 와일드카드 사용 가능)
from glob import glob

target = '*.txt'    # 확장자가 'txt'인 파일만 target으로 지정
ret = glob(target)  # 확장자가 'txt'인 파일만 얻어 리스트로 리턴
print(ret)
```

```python
# 현재 작업 디렉토리 및 하위 디렉토리에 존재하는 모든 파일과 디렉토리를 리스트로 리턴
target = '**'  # '**' => 현재 작업 디렉토리에 있는 모든 파일 및 하위 디렉토리를 의미
ret = glob(target, recursive = True)  # recursive = True => target으로 지정된 디렉토리와 그 하위 디렉토리까지 모두 탐색해 찾은 파일과 디렉토리들을 리스트로 리턴
ret.sort()  # 디렉토리 탐색 결과 정력
for r in ret :
  print(r)  # 결과 출력
```
