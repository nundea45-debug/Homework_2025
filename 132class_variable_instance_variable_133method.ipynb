```python
# Unit 132 클래스 변수(class variable)와 인스턴스 변수(instance variable)

# 클래스 변수(class variable) => 매서드 바깥에서 정의되어 클래스 범위에서 유효한 속성이나 값을 저장하는 용도로 사용
# 인스턴스 변수(instance variable) => 'self.변수이름'으로 정의되어 객체 범위에서 유효한 속성이나 값을 저장하는 용도로 사용
```

```python
# 마법사 직업에대한 클래스
class Wizard :
  generated = 0                 # 마법사 캐릭터 생성 수, 클래스 변수로 정의  # generated = 0 => 마법사 캐릭터 생성 수를 저장하는 용도의 클래스 변수

  def __init__ (self, name) :   # 클래스 생성자                             # Wizard 클래스 생성자(클래스가 인스턴스화 될 때 자동으로 실행되는 특별한 메서드)
    Wizard.generated += 1       # 캐릭터가 생성되면 클래스 변수 1증가       # generated => Wizard 클래스 생성자에서 1을 증가(클래스 변수는 '클래스 이름.클래스변수'로 처리)
    self.name = name            # 캐릭터 이름 설정                          # 인스턴스 변수("self.변수이름"으로 정의)
    self.level = 1              # 최초 레벨 1로 설정                        # 인스턴스 변수
    self.hp = 100               # 최초 체력 10으로 설정                     # 인스턴스 변수
    self.initAbility()          # 능력치 초기화

  def initAbility(self) :       # 능력치 초기화 메서드
    self.strength = 1
    self.offense_power = 1
    self.intelligence = 10
    self.mentality = 10
    self.agility = 1
```

```python
# 마법사 캐릭터 생성 수와 캐릭터 이름 출력
player1 = Wizard('FirePlayer')
print(f'마법사생성 수 : {player1.generated}\n캐릭터 이름 : {player1.name}')
# player1.generated => player 객체의 인스턴스 변수 generated를 참조 해야함 => but Wizard 객체에서 인스턴스 변수 generated 존재 X -> 클래스 변수 generated를 참조
```

```python
# 마법사 생성 수와 캐릭터 이름 출력
player2 = Wizard('GodGo')
print(f'마법사생성 수 : {player2.generated}\n캐릭터이름 : {player2.name}')  # 클래스 변수 generated를 참조 => 값이 2가 됨
```

```python
# player1.level = 2로 설정 후 player2.level의 값과 출력
player1.level = 2
print(f'player1.level = {player1.level}, player2.level = {player2.level}')  # 인스턴스 변수인 level은 객체 범위에서 유효 => player1과 player2의 level은 서로 독립적 변수
```

```python
# 인스턴스 메서드(Instance Method)
# 인스턴스 메서드 => 클래스 내의 인스턴스 멤버(인스턴스 변수, 인스턴스 메서드)들을 다루고 처리하는 용도
```

```python
# 클래스 메서드(Class Method)
# 클래스 메서드 => 인스턴스 멤버에 대한 접근 필요 없이 클래스 변수를 다루고 처리하는 용도로 사용
```

```python
# 클래스 내의 클래스 메서드 정의
@classmethod  # 클래스 메서드는 데코레이터(@)를 이용해 정의
def 메서드이름(cls, 매개변수1, 매개변수2, ...) :  # 클래스 메서드의 첫 번째 매개변수는 cls로 시작 => 클래스를 지시하는 참조자 (매개변수 필요시 두 번째 매개변수부터 정의)
  코드                                           # 클래스 메서드는 cls 매개변수를 이용해 클래스 안의 클래스 변수에 대한 접근 가능
```

```python
# 정적 메서드(Static Method)
# 정적 메서드 => 인스턴스 멤버나 클래스 변수와 관련없는 작업을 위해 사용되는 메서드
```

```python
# 클래스 내의 정적 메서드 정의
@staticmethod                               # 정적 메서드는 테코레이터(@)를 이용해 정의
def 메서드이름(매개변수1, 매개변수2, ...) :  # 정적 메서드의 파라미터는 self, cls 없이 바로 매개변수로 시작
  코드                                      # 인스턴스 참조자인 self가 매개변수로 존재 X => 클래스 안에서 인스턴스 멤버에 대한 접근 불가
```

```python
# 정적 메서드 호출 (클래스 메서드의 호출 방법과 동일)
클래스 이름.메서드이름(인자1, 인자2, ...)
```

```python
# 예제
class Wizard :
  generated = 0  # 클래스 변수 generated를 0으로 초기화

  # 인스턴스 메서드
  def setFireballLevel(self, power = 1) :  # 인스턴스 메서드 setFireballLevel()을 정의
    self.fireballLevel = power             # 인스턴스 변수 fireballLevel의 값을 인자로 입력된 power로 설정
    print(f'화염구 강도를 {power}로 설정합니다.')

  # 클래스메서드
  @classmethod
  def addGenerated(cls, count) :  # 클래스 메서드 addGenerated()를 정의
    cls.generated += count        # 클래스 변수 generated를 인자로 입력된 count로 설정, 클래스 참조자 cls를 이용해 클래스 변수에 접근하여 처리
    print(f'생성된 총 캐릭터 수 : {cls.generated}')

  # 정적 메서드
  @staticmethod
  def static1() :  # 정적 메서드 static1()을 정의 => static1은 매개변수 없이 호출되면 단순 메시지만 출력
    print('Wizard 클래스에서 정의된 정적 메서드입니다.')
```

```python
player1 = Wizard()           # Wizard 객체 player1을 생성
player1.setFireballLevel(8)  # Wizard 객체 player1에서 인스턴스 메서드의 호출 => player1.setFireballLevel(8)과 같이 호출
Wizard.addGenerated(2)       # 클래스 메서드 addGenerated() 호출 => 클래스 이름을 이용해 참조 (Wizard.addGenerated()는 Wizard 클래스의 클래스 메서드 addGenerated()를 호출)
Wizard.static1()             # 정적 메서드 static1()을 호출 => 클래스 이름을 이용해 참조 (Wizard.static1()은 Wizard 클래스의 정적 메서드 static1()을 호출)
```
