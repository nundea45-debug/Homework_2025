```python
# Unit 136 클래스 상속(inheritance)

# 클래스 상속(inheritance) => A 클래스가 가진 모든 속성이나 메서드를 B 클래스로 물려줌 -> B 클래스에서 A 클래스의 속성 및 메서드를 활용할 수 있게 함
# A 클래스 = 부모클래스, 슈퍼클래스(상속을 해 줌) / B 클래스 = 자식클래스, 서브클래스(상속을 받음)
```

```python
# 클래스 상속 방법 클래스 정의
class 자식클래스(부모클래스) :
```

```python
# 게임 캐릭터를 위한 클래스 정의
class Character :
  def __init__(self) :  # 생성자에서 캐릭터의 공통 능력치를 1로 초기화
    self.strength = 1
    self.offense_power = 1
    self.intelligence = 1
    self.mentality = 1
    self.agility = 1
```

```python
# 마법사를 위한 클래스 정의
#Character 클래스에는 캐릭터 능력치를 초기화하는 생성자만 구현 -> Wizard 클래스는 Character 클래스의 생성자와 초기화되는 능력치를 상속 받음
#-> Wizard 객체가 생성되는 시점에 상속받은 Character 클래스의 생성자 실행
class Wizard(Character) :  # Character 클래스로부터 상속 받아 Wizard 클래스를 정의
  def setBasic(self, name=None, level = 1, hp = 100) :  # Character 클래스에 없는 메서드 setBasic()을 정의
    self.name = name
    self.level = level
    self.hp = hp
```

```python
player1 = Wizard()                                # Wizard 객체 player1 생성
player1.setBasic('StormBaby')                     # setBasic() 메서드로 캐릭터 이름 설정
print(f'[{player1.name}] 레벨 : {player1.level}') # player1.name과 player1.level은 Wizard 클래스에서 정의한 변수
print(f'정신력 : {player1.mentality}')            # player1.mentality는 Character 클래스에서 상속받은 속성 mentality
```

```python
# Wizard 클래스에 생성자를 추가하고 setBasic() 메서드를 제거한 코드로 수정
class Wizard(Character) :
  def __init__(self, name=None) :
    self.name = name
    self.level = 1
    self.hp = 100
```

```python
# Wizard 객체 player1을 생성 후 주요 정보 출력
player1 = Wizard('StromBaby')
print(f'[{player1.name}] 레벨 : {player1.level}')
print(f'정신력 : {player1.mentality}')  # Wizard 객체에 mentality 속성이 X => 오류 발생 (부모클래스에 생성자가 있는데 자식클래스에서 생성자를 다시 정의했기 때문)
# 부모클래스의 생성자보다 자식클래스의 생성자가 우선(생성자의 이름이 동일시)
# => player1 객체가 생성되는 시점에 자식클래스의 생성자를 수행 -> 부모클래스의 생성자에서 정의되고 초기화되는 능력치들이 player1에 존재 X
```

```python
class Wizard(Character) :
  def __init__(self, name=None) :
    super().__init__()  # super() => 부모클래스를 의미 # 부모클래스의 생성자를 실행
    self.name = name
    self.level = 1
    self.hp = 100
```

```python
# 자식클래스에서 부모클래스의 속성이나 메서드를 직접 사용하고자 함 => 'super().메서드이름'으로 접근하거나 호출
player1 = Wizard('StromBaby')
print(f'[{player1.name}] 레벨 : {player1.level}')
print(f'정신력 : {player1.mentality}')
```

```python
# 다중상속(자식클래스가 여러개의 부모클래스로부터 상속받음) 클래스 정의
class 자식클래스(부모클래스1, 부모클래스2, ...) :
```

```python
# 다중상속 클래스의 경우 자식클래스의 생성자 정의 X => 자식클래스 객체 생성 시 '부모클래스1'의 생성자 호출 (부모클래스1의 생성자도 정의 X => 그 다음 상속 순서의 클래스 생성자를 호출)
class Parent1 :
  def __init__(self) :
    print('부모클래스1 생성자')

class Parent2 :
  def __init__(self) :
    print('부모클래스2 생성자')

class Child(Parent1, Parent2) :
  pass

obj = Child()
```

```python
# Unit 137 예외처리 1 : try~except

# 예외(exception) => 컴퓨터 프로그램이 동작 중 예상치 못한 오류로 인해 프로그램이 잘못된 동작을하거나 중단되는 경우의 상황
# 예외처리(exception handing) => 예외를 처리하는 행위
```

```python
# '파일 이름.확장자' 형식의 파일에서 파일 이름과 확장자를 분리해 출력하는 로직을 구현
fileList = ['file01.jpg', 'file02.jpg', 'file03', 'file04.jpg']
for file in fileList :
  fname = file.split('.')  # 파일 이름과 파일확장자 분리     # 'file03'에 대해 3라인을 실행한 결과 => fname = ['file03]
  print(f'파일 이름 : {fname[0]}, 파일확장자 : {fname[1]}')  # IndexError 발생 이유 => fname = ['file03']으로 fname에 없는 인덱스 1인 요소에 접근하려는 fname[1]때문
  # 확장자가 없는 file03을 처리 => IndexError 예외가 발생
```

```python
# <try~excpet>를 추가한 코드
for file in fileList :
  try :
    fname = file.split('.')  # 파일 이름과 파일확장자 분리
    print(f'파일 이름 : {fname[0]}, 파일 확장자 : {fname[1]}')
  except :
    print(f'ERROR!! <{file}> 처리중 예외 발생!')
# 예외가 발생하더라도 프로그램 종료 X => 예외 발생에 대한 메시지 출력후 프로그램 동작을 계속 수행
# <try~except> => 프로그램에서 논리적 오류나 예외가 발생할 확률이 높은 곳에 사용, 예외 발생 시 적절한 예외처리, 프로그램 중단 없이 계속 동작하도록 함
```

```python
# try~excpet 적용 방법
try :
  잠재적으로 오류 및 예외가 발생할 수 있는 코드
except :
  예외가 발생했을 때 예외를 처리하는 코드
```
