```python
# Unit 118 딕셔너리의 모든 값 조회하기 : d.values()

# d.values() => 딕셔너리 d의 모든 값을 조회할 수 있는 뷰 객체(view object)를 생성 (d의 내용 변경 시 뷰 객체도 변경된 내용이 반영됨)
```

```python
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
levels = greats.values()  # 딕셔너리 greats의 모든 값을 조회할 수 있는 뷰 객체 생성 (levels에 할당)
print(list(levels))        # greats의 값들로만 이루어진 리스트 출력
```

```python
# greats와 뷰 객체 levels가 연관되 있음을 보여주는 예
greats['조운'] = 97
print(list(levels))  # greats에서 키가 '조운'인 요소의 값을 97로 변경한 내용이 levels에도 반영됨
```

```python
# Unit 119 딕셔너리에서 모든 키의 순서를 거꾸로 하기: reversed(d)

# reversed() => 딕셔너리에 입력시 딕셔너리의 모든 키의 순서를 거꾸로 한 이터레이터를 리턴
```

```python
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
levels = reversed(greats)  # 딕셔너리 greats의 모든 키의 순서를 거꾸로 한 이터레이터를 리턴
print(list(levels))
```
