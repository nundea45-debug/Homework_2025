```python
# Unit 125 함수의 매개변수(parameter)와 인자

# 함수의 매개변수(parameter) => 함수를 정의할 때 함수로 전달할 값이 입력될 부분에 명시되는 변수 (ex. def func(param) => param이 함수 func의 매개변수)
```

```python
# 디폴트 인자를 이용한 인자 전달 방법 예시
def setLevel(character, level1 = 1) :  # setLevel()을 정의  # level의 디폴트 인자로 1을 설정 => setLevel() 호출 시 두 번째 인자가 입력 X -> level을 디폴트 인자 1로 하고 함수 실행
  print(f'{character}의 레벨 : {level1}')
  return
```

```python
setLevel('유비')  # level의 인자 없이 호출 -> level의 값을 기본값 1로 설정 후 함수 실행
setLevel('제갈량', 10)  # level의 인자로 10을 지정해 호출 -> 함수 호출 시 지정한 인자 10이 level의 값으로 설정
```

```python
# 디폴트 인자는 반드시 함수 마지막 매개변수로 지정 -> 매개변수가 일반 매개변수보다 앞에 있을 시 => SyntaxError 오류 발생
def setLevel(level = 1, character) :
```

```python
# 키워드 인자를 이용한 인자 전달 방법 예시
def setLevel(character, level) :  # setLevel()을 정의
  print(f'{character}의 레벨 : {level}')
  return
```

```python
setLevel(level = 5, character = '장비')  # 함수의 매개변수에 직접 값을 대입하여 인자를 전달하는 방식 (인자의 순서 상관 X)
```

```python
# 가변 인자를 이용한 인자 전달 방법 예시
def setLevel(*args) :  #  setLevel()을 호출할 때 입력되는 인자들을 튜플로 묶어 args로 전달 (호출할 때 입력되는 인자 개수와 무관하게 동작)
  print(args)
  return
```

```python
setLevel('관우', 7)  # 인자의 개수가 2개
setLevel('조운', 'named', 10)  # 인자의 개수가 3개  # 매개변수 args는 호출시 입력되는 가변 인자를 튜플로 전달받음
```

```python
# 가변 키워드 인자를 이용한 인자 전달 방법 예시
# ** => 함수의 매개변수 이름 앞에 붙여 전달 받을 인자가 가변적 키워드 인자임을 알림
def setLevel(**kwargs) :  # setLevel()을 호출할 때 입력되는 키워드 인자들을 요소로하는 딕셔너리를 kwargs로 전달 받음 (setLevel()을 호출할 때 입력되는 키워드 인자 개수와는 무관하게 동작)
  print(kwargs)
  return
```

```python
setLevel(character = '조운', level = 10)
setLevel(character = '황충', category = 'named', level = 5)  # setLevel(**kwargs) => 호출시 입력된 키워드 인자들을 kwargs에 딕셔너리로 전달
```

```python
# 명확한 인자와 가변 인자가 함께 존재하는 경우 함수 정의 및 호출 방법 예시
def setLevel(character, level, *args) :  # character, level => 명확하게 정의된 매개변수, args를 통해 추가적인 인자를 전달 받음 -> '기타 특성'으로 출력
  print(f'장수 : {character}, 레벨 : {level}')
  if len(args) != 0 :
    print(f'기타 특성 : {args}')
  return
```

```python
setLevel('조조', 3)  # args에 전달할 인자가 X => character와 level만 출력
```

```python
setLevel('유비', 3, 'named', 'leader')  # args에 ('named', 'leader')가 인자로 전달
```

```python
# 명확한 인자와 가변 키워드 인자가 함께 존재하는 경우 함수 정의 및 호출 방법 예시
def setLevel(character, level, **kwargs) :  # character, level => 명확하게 정의된 매개변수, kwargs를통해 추가적인 키워드 인자를 딕셔너리로 전달 받음 -> "매개변수 : 값"으로 출력
  print(f'장수 : {character}, 레벨 : {level}')
  if len(kwargs) != 0 :
    for k, v in kwargs.items() :
      print(f'{k} : {v}')
    return
```

```python
setLevel('조조', 3)  # kwargs에 전달할 키워드 인자가 X => character와 level만 출력
```

```python
setLevel('유비', 3, category = 'named', position = 'leader')  # kwargs에 딕셔너리 {category = 'naemd', position = 'leader'}가 인자로 출력
```
