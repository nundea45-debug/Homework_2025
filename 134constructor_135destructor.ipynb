```python
# Unit 134 클래스 생성자(Constructor)

# 클래스 생성자(Constructor) => 클래스의 인스턴스 객체가 생성되는 시점에 자동으로 실행되는 특별한 메서드
```

```python
# 클래스 생성자 정의
def __init__(self, 매개변수1, 매개변수2, ...) :
  코드
```

```python
# Wizard 클래스에서 클래스 생성자 정의
class Wizard :
  def __init__(self) :  # 매개변수 없이 정의
    print('마법사 캐릭터가 생성되었습니다.')
```

```python
# Wizard 객체 생성
player1 = Wizard()  # Wizard 클래스 생성자가 인자를 가지지 X => 클래스 객체 생성은 Wizard()와 같이 인자 없이 수행, Wizard 객체가 생성될 때 클래스 생성자가 자동 실행
```

```python
# 인자가 있는 클래스 생성자 정의
class Wizard :
  def __init__(self, name=None) :  # Wizard 클래스 생성자가 name이라는 매개변수를 1개 가짐 => Wizard 객체 생성시 name에 입력할 인자를 지정해야함
    self.name = name               # Wizard 클래스의 생성자는 인스턴스 변수 'self.name = name'으로 설정
    print(f'마법사 이름이<{self.name}>인 캐릭터가 생성되었습니다.')  # 설정한 마법사 이름을 출력
```

```python
player1 = Wizard('StormBaby')  # Wizard 객체 player1을 생성할 때 생성자에 전달할 인자로 'StormBaby'를 입력
```

```python
# Unit 135 클래스 소멸자(Destructor)

# 클래스 소멸자(Destructor) => 클래스의 인스턴스 객체가 메모리에서 제거되는 시점에 자동으로 실행되는 특별한 클래스 메서드
```

```python
# 클래스 소멸자 정의
def __del__(self) :
  코드
```

```python
# Wizard 클래스에 소멸자 코드 추가
class Wizard :
  def __init__(self, name=None) :
    self.name = name
    print(f'마법사 이름이 <{name}>인 캐릭터가 생성되었습니다.')

  def __del__(self) :
    print('Wizard 객체가 메모리에서 제거됩니다.')
```

```python
# Wizard 객체 생성 후 del 키워드를 이용해 생성한 객체를 제거하는 코드
player1 = Wizard('StormBaby')
del player1  # player1 객체를 메모리에서 제거함
```
