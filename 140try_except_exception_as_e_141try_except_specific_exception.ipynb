```python
# Unit 140 예외처리 4 : try~except Exception as e

# 코드에서 예외 발생 시 이에 대한 자세한 내용을 파악하는 것이 중요
# <try~except Exception as e>로 둘러 싸인 코드에서 오류 발생 => 발생한 오류는 Exception으로 참조 가능
```

```python
intList = range(10)
idx = 10
try :
  print(intList[idx])    # 범위를 벗어난 인덱스 10으로 intList의 요소에 접근하려 함 => 오류 발생
except Exception as e :  # <except Exception as e : > => 발생한 예외 Exception을 e로 둔다는 의미(Python은 모든 예외에 대해 Exception 클래스로 미리 정의)
  exceptionClass = e.__class__.__name__                           # 발생한 예외 클래스의 이름을 가져옴
  exceptionDetail = e.__str__()  # print(a) = print(e.__str__())  # 발생한 예외 내용을 가져옴
  print(f'예외종류 : {exceptionClass}, 예외내용 : {exceptionDetail}')
```

```python
intList = range(10)
idx = 10
try :
  print(intList[idx])
except Exception as e :
  exceptionClass = e.__class__.__name__
  exceptionDetail = e.__str__()  # print(a) = print(e.__str__())
  print(f'예외종류 : {exceptionClass}, 예외내용 : {e}') # => 이와 같이 수정해도 동일한 결과 출력
```

```python
# Unit 141 예외처리 5 : try~except 특정예외

# try~except 특정예외 => 코드에서 특정 예외가 발생했을 때만 except에서 처리할 수 있는 방법 -> except 다음에특정 예외 클래스를 명시
try :
  잠재적으로 오류 및 예외가 발생할 수 있는 코드
except 예외클래스 1 :
  예외1이 발생했을 때 예외를 처리하는 코드
except 예외클래스 2 :
  예외2가 발생했을 때 예외를 처리하는 코드
```

```python
intList = range(10)
try :
  idx = int(input('숫자를 입력하세요 :'))  # int()의 인자로 숫자가 아닌 문자 입력 => ValueError 발생
  print(intList[idx])
except ValueError :                       # 에러가 ValueError일 시 => 6라인 실행
  print('입력한 내용이 숫자가 아닙니다.')
except IndexError as e :
  print(f'인덱스 오류가 발생 : {e}')
```
