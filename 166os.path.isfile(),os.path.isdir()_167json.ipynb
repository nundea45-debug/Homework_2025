```python
# Unit 166 파일인지 디렉토리인지 체크 : os.path.isfile(), os.path.isdir()

# os.path.isfile(target) => target이 파일 -> True / 파일 X -> False를 리턴 (target => 절대경로, 상대경로로 지정된 값)
# os.path.isdir(target) => target이 디렉토리 -> True / 디렉토리 X -> False를 리턴 (target => 절대경로, 상대경로로 지정된 값)
# isfile(target)과 isdir(target)은 리턴값만 다름 => target이 파일인지 디렉토리인지 확인한다는 목적은 동일
```

```python
# 현재 작업 디렉토리에 존재하는 파일과 디렉토리를 구분하여 오름차순으로 출력하는 코드
from os import listdir
from os.path import isdir

fileList = []          # 파일과 디렉토리 목록을 저장할 리스트를 정의
dirList = []
files = listdir('./')  # 현재 작업 디렉토리의 파일 리스트  # 현재 작업 디렉토리에 존재하는 파일과 디렉토리를 리스트로 얻고 files로 둠
for file in files :
  if isdir(file) :     # files의 요소 file이 디렉토리이면 '<DIR> file'을 dirList에 추가
    dirList.append(f'<DIR> {file}')
  else :               # files의 요소 file이 디렉토리가 아니면(파일이면) 'file'을 fileList에 추가
    fileList.append(file)

fileList.sort()        # fileList와 dirList를 오름차순으로 정렬
dirList.sort()

for d in dirList :     # dirList의 요소들을 모두 출력
  print(d)

for f in fileList :    # fileList의 요소들을 모두 출력
  print(f)
```

```python
# pakage.jpg 파일 내용을 읽어 딕셔너리로 저장하는 코드
import json

with open('package.json') as f :  # JSON 파일을 텍스트 읽기 모드로 오픈
      data = json.load(f)         # JSON 파일 f의 내용을 읽어 딕셔너리로 리턴
print(data)
```

```python
profile = {'name' : '조조', 'age' : 45, 'nationality' : '위나라',
           'skill' : ['집착', '배신'], 'alive' : True}   # 삼국지 '조조'의 간단한 프로파일을 저장한 딕셔너리

with open('profile.json', 'w') as f :  # 파일을 텍스트 쓰기 모드로 오픈
  json.dump(profile, f, indent = 4)  # profile을 JSON 파일 f에 저장 / indent = 4 => profile의 '키:값'쌍을 한 줄로 저장(하위 수준의 쌍이나 값들은 지정한 인덴트 들여쓰기로 한 줄로 저장)
```
