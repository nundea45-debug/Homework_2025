```python
# Unit 116 딕셔너리의 모든 요소 조회하기 : d.items()

# d.items() => 딕셔너리 d의 모든 요소들을 조회할 수 있는 뷰 객체(view object)를 생성 (d의 내용 변경 시 뷰 객체도 변경된 내용 반영)
```

```python
greats = {'유비':89, '관우':95, '장비':91, '조운':94}
generals = greats.items()  # 딕셔너리 greats의 모든 요소를 조회할 수 있는 뷰 객체를 생성 -> generals에 할당
print(list(generals))      # 뷰 객체 generals를 리스트로 변환해 출력 => greats의 요소 key:val이 튜플로 되있음
```

```python
# 딕셔너리 greats와 뷰 객체 generals가 연관됨을 보여주는 예
del greats['관우']  # greats에서 '관우':95를 삭제
print(list(generals))  # generals도 변경된 내용이 반영된 결과 출력
```

```python
# Unit 117 딕셔너리의 모든 키 조회하기 : d.keys()

# d.keys() => 딕셔너리 d의 모든 키들을 조회할 수 있는 뷰 객체(view object)를 생성 (d의 내용 변경 시 뷰 객체도 변경된 내용 반영)
```

```python
greats = {'유비' : 89, '관우':95, '장비':91, '조운':94}
general_names = greats.keys()  # 딕셔너리 greats의 모든 키를 조회할 수 있는 뷰 객체 생성 -> general_names로 둠
print(list(general_names))  # 뷰 객체 general_names를 리스트로 변환 해 출력 => greats의 키로만 구성된 리스트임
```

```python
# 딕셔너리 greats와 뷰 객체 general_names가 연관됨을 보여주는 예
del greats['관우']  # greats에서 '관우'가 키인 요소 삭제
for i, name in enumerate(general_names, start = 1):
  print(f'남아 있는 촉나라 장수[{i}] : {name}')
```
